```python
from sympy import mod_inverse

def curve_checker(a, b, p, x1, y1):
    LH = x1**3 + a * x1 + b
    RH = y1**2

    LH_mod = LH % p
    RH_mod = RH % p

    if LH_mod == RH_mod:
        return True
    else:
        return False

def point_addition(a, b, p, x1, y1, x2=None, y2=None):
    if x2 is None and y2 is None:
        x2, y2 = x1, y1

    if (x1 == x2) and (y1 == y2):
        numerator = 3 * x1**2 + a
        denominator = 2 * y1
    else:
        numerator = y2 - y1
        denominator = x2 - x1


    if denominator == 0:
        return float('inf'), float('inf')

    denominator_inv = mod_inverse(denominator, p)
    λ = (numerator * denominator_inv) % p


    x3 = (λ**2 - x1 - x2) % p
    y3 = (λ * (x1 - x3) - y1) % p

    return (x3, y3)

def find_group_order(a, b, p, x1, y1):
    x, y = x1, y1
    count = 1
    print(f"Value of {count}p is ({x},{y})")
    while True:
        x, y = point_addition(a, b, p, x1, y1, x, y)

        if x == float('inf') and y == float('inf'):
            break
        print(f"Value of {count+1}p is ({x},{y})")
        count += 1

        if x == x1 and y == y1:
            break

    return count

a, b, p = 2, 2, 17
x1, y1 = 5, 1

if curve_checker(a, b, p, x1, y1):
    # Find the group order
    group_order = find_group_order(a, b, p, x1, y1)
    print(f"The group order of the point P({x1}, {y1}) is {group_order}.")
else:
    print(f"The point P({x1}, {y1}) is not on the curve.")
```
