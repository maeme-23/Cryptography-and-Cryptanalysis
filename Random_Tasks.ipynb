#Task 3

Image: https://drive.google.com/file/d/1ea2M_L4RsJLUwGnj8LBaErxiek3CbAox/view?usp=sharing

Message: **birds_of_a_same_feather_flock_together**

Hide the message inside the green layer of the image and extract the message again to verify

```python
# Your code here

from PIL import Image
```

```python
def hide_message(image_path, message):
    img = Image.open(image_path)  # Use the image_path argument here
    width, height = img.size

    binary_message = "".join(format(ord(i), "08b") for i in message)
    max_message_length = (width * height) // 8

    if len(binary_message) > max_message_length:
        raise ValueError("Message too long to be hidden in the given image")

    # Convert image to RGB mode to ensure consistent pixel format
    img = img.convert("RGB")

    new_img = img.copy()  # Create a copy of the original image instead of new grayscale

    message_index = 0

    for x in range(width):
        for y in range(height):
            r, g, b = img.getpixel((x, y)) # Get individual RGB values

            if message_index < len(binary_message):
                # Modify the green channel only
                g_binary = format(g, "08b")
                g_binary = g_binary[:-1] + binary_message[message_index]
                g = int(g_binary, 2)
                message_index += 1

            # Update the pixel with the modified green channel
            new_img.putpixel((x, y), (r, g, b))

    new_img.save("hidden.png")  # Save the modified image
```

```python
hide_message("/content/bird.jpg", "birds_of_a_same_feather_flock_together")
```

```python
def extract_message(image_path):
    img = Image.open(image_path)
    width, height = img.size

    binary_message = ''
    for x in range(width):
        for y in range(height):
            r, g, b = img.getpixel((x, y))  # Get RGB values as individual components
            binary_message += format(g, '08b')[-1]  # Use green channel (g)

    message = ''
    for i in range(0, len(binary_message), 8):
        byte = binary_message[i:i + 8]
        if byte == '11111111':  # Check for padding (if used)
            break
        message += chr(int(byte, 2))

    print(message)
```
